```python
import numpy as np        
import pandas as pd       
import matplotlib.pyplot as plt  
```

```python
distances = np.array([5, 10, 15, 20, 25, 30, 35])
angles_deg = np.array([90, 34, 87, 54, 0, 15, 45])

prob_distance_noob = 1 / (1 + np.exp(0.25 * distances - 1.5))
prob_distance_pro = 1 / (1 + np.exp(0.12 * distances - 1.0))

angles_rad = angles_deg * np.pi / 180
angle_sin_values = np.sin(angles_rad)

prob_angle_noob = 1 / (1 + np.exp(-5 * angle_sin_values + 3))
prob_angle_pro = 1 / (1 + np.exp(-8 * angle_sin_values + 3))

xG_pro = prob_distance_pro * prob_angle_pro
xG_noob = prob_distance_noob * prob_angle_noob
```

```python
plt.plot(distances, prob_distance_noob, 'o-', color='red', label='Noob')
plt.plot(distances, prob_distance_pro, 'o-', color='green', label='Pro')
plt.title('Goal Probability based on Distance and Skill Level')
plt.xlabel('Distance from goal (meters)')
plt.ylabel('Probability (0-1)')
plt.grid(True)
plt.legend()
plt.show()
```
